```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
opencv_logo = "../../image4labs/openCV_logo_white.png"

img = cv.imread(opencv_logo)
assert img is not None, "file could not be read, check with os.path.exists()"


def make_plot(img, dst, title1, title2):
    plt.subplot(121), plt.imshow(img), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(dst), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
kernel = np.ones((5, 5), np.float32)/25
dst = cv.filter2D(img, -1, kernel)

make_plot(img, dst, 'Original', 'Averaging')
```

```python
blur = cv.blur(img, (5, 5))

make_plot(img, dst, 'Original', 'Blurred')
```

```python
blur = cv.GaussianBlur(img, (5, 5), 0)

make_plot(img, dst, 'Original', 'Blurred')
```

```python
median = cv.medianBlur(img, 5)

make_plot(img, dst, 'Original', 'Median')
```

```python
blur = cv.bilateralFilter(img, 9, 75, 75)

make_plot(img, dst, 'Original', 'Blurred')
```
