```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
state = '../../image4labs/messi.jpg'
img = cv.imread(state)[..., ::-1]
assert img is not None, "file could not be read, check with os.path.exists()"

plt.imshow(img)
```

- каналы = [0,1], поскольку нам необходимо обработать как плоскость H, так и плоскость S.
- ячейки = [180,256] 180 для плоскости H и 256 для плоскости S.
- диапазон = [0,180,0,256] Значение оттенка находится в диапазоне от 0 до 180, а насыщенность — в диапазоне от 0 до 256.


```python
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
hist = cv.calcHist([hsv], [0, 1], None, [180, 256], [0, 180, 0, 256])
hist = cv.normalize(hist, hist, alpha=0, beta=1, norm_type=cv.NORM_MINMAX)
plt.imshow(hist, interpolation='nearest')
plt.show()
```

Ось X показывает значения S, а ось Y — оттенок.
